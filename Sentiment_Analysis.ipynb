```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
df = pd.read_csv('/content/Twitter_Data.csv')
```

```python
print("Dataset Info:")
print(df.info())
```

```python
print("\nFirst Few Rows:")
print(df.head())
```

```python
print("\nSentiment Distribution:")
print(df['category'].value_counts())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='category', data=df)
plt.title('Sentiment Distribution')
plt.show()
```

```python
from wordcloud import WordCloud

positive_text = ' '.join(df[df['category'] == 'positive']['clean_text'])
negative_text = ' '.join(df[df['category'] == 'negative']['clean_text'])
```

```python
if 'positive' in df['category'].astype(str).unique():
    # Extract positive text
    positive_text = ' '.join(df[df['category'] == 'positive']['clean_text'])

    # Display word cloud for positive sentiment
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    wordcloud_positive = WordCloud(width=400, height=400, random_state=42).generate(positive_text)
    plt.imshow(wordcloud_positive, interpolation='bilinear')
    plt.title('Positive Sentiment Word Cloud')
    plt.axis('off')
else:
    print("No positive sentiments in the dataset.")

plt.show()
```

```python
if 'negative' in df['category'].astype(str).unique():
    # Extract negative text
    negative_text = ' '.join(df[df['category'] == 'negative']['clean_text'])

    # Display word cloud for negative sentiment
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 2)
    wordcloud_negative = WordCloud(width=400, height=400, random_state=42).generate(negative_text)
    plt.imshow(wordcloud_negative, interpolation='bilinear')
    plt.title('Negative Sentiment Word Cloud')
    plt.axis('off')
else:
    print("No negative sentiments in the dataset.")

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='category', data=df)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['category'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Sentiment Distribution')
plt.ylabel('')
plt.show()
```
